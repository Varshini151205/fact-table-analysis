```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('fact_table.csv', encoding='ISO-8859-1')

# Optional: Rename column if there's a typo
df.rename(columns={'coustomer_key': 'customer_key'}, inplace=True)

# Countplot: Frequency of quantities ordered
plt.figure(figsize=(8, 4))
sns.countplot(x='quantity', data=df)
plt.title('Quantity Distribution (Count)')
plt.xlabel('Quantity')
plt.ylabel('Count')
plt.tight_layout()
plt.show()



```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('fact_table.csv', encoding='ISO-8859-1')

# Optional: Rename column if there's a typo
df.rename(columns={'coustomer_key': 'customer_key'}, inplace=True)
# Histogram: Distribution of total_price
plt.figure(figsize=(8, 4))
sns.histplot(df['total_price'], bins=20, kde=True)
plt.title('Total Price Distribution')
plt.xlabel('Total Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('fact_table.csv', encoding='ISO-8859-1')

# Optional: Rename column if there's a typo
df.rename(columns={'coustomer_key': 'customer_key'}, inplace=True)
# Boxplot: total_price by store_key
plt.figure(figsize=(8, 4))
sns.boxplot(x='store_key', y='total_price', data=df)
plt.title('Total Price per Store')
plt.xlabel('Store')
plt.ylabel('Total Price')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('fact_table.csv', encoding='ISO-8859-1')

df.rename(columns={'coustomer_key': 'customer_key'}, inplace=True)
# Line plot: quantity trend (index-based)
plt.figure(figsize=(8, 4))
df['quantity'].plot(kind='line', marker='o')
plt.title('Quantity Trend Over Time (by index)')
plt.xlabel('Index')
plt.ylabel('Quantity')
plt.tight_layout()
plt.show()
```
